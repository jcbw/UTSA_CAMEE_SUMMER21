> After loading our data, we can now use the information to plot our profiles. Here, we plot the temperature-depth profiles for Site 2&5. As expected, we see a change in the temperature with increasing depth. 

```python
# Dropping entries where temp = 0
site2 = site2[site2['degree_c'] != 0]
site5 = site5[site5['degree_c'] != 0]
```

```python
plt.figure(figsize=(10,5))
plt.plot(site2.degree_c,site2.depth_cm,'k--', label = 'Site 2', alpha = 0.60)
plt.plot(site5.degree_c,site5.depth_cm,'m--', label = 'Site 5', alpha = 0.70)
plt.ylabel('Depth/(cm)')
plt.xlabel('Temperature/($^\circ$C)')
plt.ylim(max(site5['depth_cm']), min(site5['depth_cm']))
plt.legend(loc='best')
plt.show()
```

> Earlier, we mentioned the use of numpy for basic mathematical functions. Here's an example of that. We display the maximum, minimum, mean and standard deviation of the salinity and depth values from our first core dataset.

```python
np.max(s2c1.salinity_ppt),np.min(s2c1.salinity_ppt),np.mean(s2c1.salinity_ppt),np.std(s2c1.salinity_ppt)
```

```python
plt.figure(figsize=(9,12))
plt.plot(s2c1.salinity_ppt,s2c1.depth_cm,'k-.',alpha = 0.60, label='site 2 - core 1')
plt.plot(s4c1.salinity_ppt,s4c1.depth_cm ,'m-*',alpha = 0.60, label='site 4 - core 1')
plt.plot(s4c2.salinity_ppt,s4c2.depth_cm ,'r-.', alpha = 0.60, label='site 4 - core 2')
plt.plot(s7c1.salinity_ppt,s7c1.depth_cm ,'g-*',alpha = 0.60, label='site 7 - core 1')
plt.plot(s22c1.salinity_ppt,s22c1.depth_cm ,'b-.',alpha = 0.60, label='site 22 - core 1')
plt.plot(s23c1.salinity_ppt,s23c1.depth_cm ,'y-*',alpha = 0.60, label='site 23 - core 1')
plt.plot(s23c3.salinity_ppt,s23c3.depth_cm ,'c-*', alpha = 0.60, label='site 23 - core 3')
plt.ylabel('Depth/(cm)')
plt.xlabel('Salinity/(ppt)')
plt.ylim(250,0)
plt.legend(loc='lower right')
plt.grid('on')
plt.show()
```
